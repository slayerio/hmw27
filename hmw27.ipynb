```python
import numpy as np
import pandas as pd
```

```python
mobile = pd.read_csv(r'C:\Users\User\Downloads\mobile.csv')
mobile
```

```python
mobile['first name'] = mobile['Customer Name'].apply(lambda name: name.split()[0] if (name, str) else '')
mobile['Phone Company'] = mobile['Phone Number'].apply(lambda x: 'Pelephone' if x[:3] == '050'
                                                       else 'Golan' if x[:3] == '051'
                                                       else 'Selcum' if x[:3] == '052' 
                                                      else 'Motorolla' if x[:3] == '053'
                                                       else 'Partner') 
mobile['User kind'] = mobile.apply(lambda x: 'Talker' if x['Talked Minutes'] > x['Internet Minutes'] else 'Surfer', axis=1)
# mobile['User kind'] = np.vectorize(lambda x, y: 'Talker' if X > y else 'Surfer')(mobile['Talked Minutes'],mobile['Internet Minutes'])
mobile
```
